```python
# calculator with except and try for eror handling 

print("""
1. Addition
2. Subtraction
3. Multiplication
4. Division
5.Exit
""")

try:
    choice = input("Enter the number: ")
    if choice == "5":
        print("Exiting Goodbye!")


    elif choice == "1":
        try:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            print("The output is:", num1 + num2)
        except ValueError:
            print("Error: Please enter valid numbers.")

    elif choice == "2":
        try:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            print("The output is:", num1 - num2)
        except ValueError:
            print("Error: Please enter valid numbers.")

    elif choice == "3":
        try:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            print("The output is:", num1 * num2)
        except ValueError:
            print("Error: Please enter valid numbers.")

    elif choice == "4":
        try:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            if num2 == 0:
                print("Error: Division by zero is not allowed.")
            else:
                print("The output is:", num1 / num2)
        except ValueError:
            print("Error: Please enter valid numbers.")

    else:
        print("The input is invalid.")

except Exception as e:
    print("Unexpected error occurred:", e)
```

```python
# Movie Ticket Booking Program with Age Verification

print("=== Welcome to the Movie Ticket Booking System ===\n")

# Movie list with minimum age requirement
movies = {
    1: ("The Lion King", 0),
    2: ("Avengers: Endgame", 13),
    3: ("John Wick", 18),
    4: ("Deadpool", 18)
}

while True:
    print("Available Movies:")
    for num, (movie, age_req) in movies.items():
        print(f"{num}. {movie} (Age {age_req}+)")

    choice = int(input("\nEnter the movie number you want to watch: "))

    if choice in movies:
        movie_name, min_age = movies[choice]
        print(f"\nYou selected: {movie_name}")
    else:
        print("Invalid movie number. Please try again.\n")
        continue

    age = int(input("Enter your age: "))
    is_allowed = age >= min_age

    if not is_allowed:
        print(f"You are not allowed to watch {movie_name}. Minimum age required is {min_age}.\n")
        continue

    print("Age verification successful.")

    while True:
        num_tickets = int(input("Enter number of tickets you want to book: "))
        
        if num_tickets > 0:
            print(f"\nSuccessfully booked {num_tickets} ticket(s) for {movie_name}.")
            print("Enjoy your movie.\n")
            break
        else:
            print("Please enter a valid number of tickets.\n")

    again = input("Do you want to book another movie? (yes/no): ").lower()
    if again != "yes":
        print("\nThank you for using the Movie Ticket Booking System. Goodbye.")
        break
```

```python
import pygame


pygame.init()


screen = pygame.display.set_mode((600, 600))
pygame.display.set_caption("X O Game")


WHITE = (255, 255, 255)
BLACK = (0, 0, 0)


SIZE = 200
board = [["", "", ""],
         ["", "", ""],
         ["", "", ""]]

player = "X"
font = pygame.font.Font(None, 100)


run = True
while run:
    
    screen.fill(WHITE)
    
    
    for x in range(1, 3):
        pygame.draw.line(screen, BLACK, (x*SIZE, 0), (x*SIZE, 600), 5)
        pygame.draw.line(screen, BLACK, (0, x*SIZE), (600, x*SIZE), 5)
    
   
    for row in range(3):
        for col in range(3):
            if board[row][col] != "":
                text = font.render(board[row][col], True, BLACK)
                screen.blit(text, (col*SIZE + 50, row*SIZE + 50))
    
   
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            mouseX, mouseY = event.pos
            clicked_row = mouseY // SIZE
            clicked_col = mouseX // SIZE
            if board[clicked_row][clicked_col] == "":
                board[clicked_row][clicked_col] = player
              
                player = "O" if player == "X" else "X"

    
    pygame.display.update()


pygame.quit()
```

```python
import matplotlib.pyplot as plt
import pyttsx3
import os
import webbrowser
import speech_recognition as sr
import pywhatkit

speaker = pyttsx3.init('sapi5')
mic = sr.Recognizer()

speaker.say("Welcome to Bixby")
speaker.runAndWait()

with sr.Microphone() as source:
    print("Start speaking...")
    mic.adjust_for_ambient_noise(source, duration=1)

    try:
        voice = mic.listen(source, timeout=5, phrase_time_limit=5)
        text = mic.recognize_google(voice)
        command = text.lower()
        print("You said:", command)

    except sr.WaitTimeoutError:
        print("No speech detected.")
        command = ""
    except sr.UnknownValueError:
        print("Sorry, I could not understand.")
        command = ""
    except sr.RequestError:
        print("Speech service down.")
        command = ""

# ---------------- COMMANDS ----------------

if "notepad" in command or "note pad" in command:
    print("Opening Notepad...")
    os.system("notepad")

elif "youtube" in command or "you tube" in command or "yt" in command:
    print("Opening YouTube...")
    webbrowser.open("https://youtube.com")

elif "spotify" in command or "spot fy" in command or "spootify" in command:
    print("Opening Spotify...")
    try:
        os.system("start spotify")
    except:
        webbrowser.open("https://open.spotify.com")
        print("Spotify Desktop not found — opening web player.")

elif "whatsapp" in command or "send message" in command:
    print("Sending WhatsApp message...")
    try:
        pywhatkit.sendwhatmsg_instantly(
            "+919548892134",
            "hi i am from vgu",
            wait_time=10,
            tab_close=True
        )
        print("Message sent!")
    except Exception as e:
        print("Error sending WhatsApp message:", e)

elif command != "":
    print("Command not recognized.")

speaker.stop()
```
